```python
# https://www.acs.psu.edu/drussell/Demos/waves/wavemotion.html
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

xy = np.random.uniform(-4, 4, (2000,2))
# set up parameters
w = 1/5
R = 0.2
k = 1.5
t = 1

def transverse_wave(frame):
    xy2 = np.copy(xy)   
    for i in range(len(xy2)):
        x = xy2[i][0]
        #update patical position use R cos(kx − ωt) + (1 − R) cos(kx + ωt)
        p = R*np.cos(w*frame - k*x) + (1-R)*np.cos(w*frame + k*x)
        xy2[i][1] += p
    sc.set_offsets(np.c_[xy2[:,0], xy2[:,1]])

xy2 = np.copy(xy)       
fig,ax = plt.subplots()
sc = ax.scatter(xy2[:,0], xy2[:,1], s=10, c="k")
plt.ylim(-5, 5)

ani = FuncAnimation(fig, transverse_wave, frames=6000, interval=20)

plt.show()
```
